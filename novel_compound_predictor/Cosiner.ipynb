# Similarity Calculation

A temporary dataset is contructed to perform faster cosine similarity calculations.

Each row contains the 18 decade columns for each of the compound and its 2 constituents.

Cosine similarity is represented using - 

\begin{equation} cos(\pmb x, \pmb y) = \frac {\pmb x \cdot \pmb y}{||\pmb x|| \cdot ||\pmb y||} \end{equation} .

The denominator values are pre-calculated for all values. 

```python
cosiner=compounds[precols]
cosiner['denominator']=np.sqrt((cosiner[precols]**2).sum(axis=1))
cosiner.columns=cosiner.columns+'_compound'
cosiner.reset_index(['modifier','head'],inplace=True)
currentcolumns=cosiner.columns.tolist()
#cosiner
cosiner=pd.concat([cosiner.set_index('modifier'),constituents], axis=1, join='inner').reset_index()
cosiner.drop(['presence','pattern'],axis=1,inplace=True)
cosiner['denominator']=np.sqrt((cosiner[precols]**2).sum(axis=1))
temp_columns=cosiner.columns[22:]+'_modifier'
cosiner.columns=currentcolumns+temp_columns.tolist()
cosiner=pd.concat([cosiner.set_index('head'),constituents], axis=1, join='inner').reset_index()
cosiner.drop(['presence','pattern'],axis=1,inplace=True)
cosiner['denominator']=np.sqrt((cosiner[precols]**2).sum(axis=1))
fornow=cosiner.columns.tolist()
temp_columns=cosiner.columns[-20:]+'_head'
cosiner.columns=['head']+[fornow[1]]+fornow[2:-20]+temp_columns.tolist()
cosiner.set_index(['modifier','head'],inplace=True)
cosiner.head(20)
```

# Computation of average cosine similarities for the constituents


The modifier cosine values are now added to the constituents in the form of 3 feautures -

**mean_sim_as_modifier** - The average similarity measure of the constituent as a modifier
**sim_with_modifier** - The standard deviation of the similarity of the constituent as a modifier
**presence_as_modifier** - The number of times the constituent appeared as a modifier.

```python
modifier_sim=compounds.groupby(level=['modifier'])['sim_with_modifier'].agg(['mean','std','size'])
modifier_sim.columns=['mean_sim_as_modifier','std_sim_as_modifier','presence_as_modifier']


modifier_sim.presence_as_modifier[(modifier_sim.mean_sim_as_modifier==0) & (modifier_sim.std_sim_as_modifier==0)]=0
modifier_sim.std_sim_as_modifier.fillna(0,inplace=True)
modifier_sim.std_sim_as_modifier[modifier_sim.mean_sim_as_modifier.isnull()]=np.nan
modifier_sim['presence_as_modifier']=modifier_sim["presence_as_modifier"].astype('category')
modifier_sim
```

Same as with the modifiers, the head cosine values are now added to the constituents in the form of 3 feautures -

**mean_sim_as_head** - The average similarity measure of the constituent as a head

**sim_with_head** - The standard deviation of the similarity of the constituent as a head

**presence_as_head** - The number of times the constituent appeared as a head.

```python
head_sim=compounds.groupby(level=['head'])['sim_with_head'].agg(['mean','std','size'])
head_sim.columns=['mean_sim_as_head','std_sim_as_head','presence_as_head']

head_sim.presence_as_head[(head_sim.mean_sim_as_head==0) & (head_sim.std_sim_as_head==0)]=0
head_sim.std_sim_as_head.fillna(0,inplace=True)
head_sim.std_sim_as_head[head_sim.mean_sim_as_head.isnull()]=np.nan
head_sim['presence_as_head']=head_sim["presence_as_head"].astype('category')
head_sim
```

The 6 new features are now added to the individual constituents.

```python
constituents=pd.concat([constituents,modifier_sim,head_sim],axis=1)
constituents
```
