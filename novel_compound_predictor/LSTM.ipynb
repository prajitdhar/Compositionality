```python
import pandas as pd
import numpy as np
import itertools
```

```python
constituents=pd.read_pickle("/data/dharp/compounding/datasets/constituents_CompoundAgnostic_DecadeCentric_300.pkl")
constituents.index.names=['constituent','decade']
constituents.info()
constituents.head()
```

```python
constituents_list=constituents.index.unique(level=0).tolist()
decades_list=constituents.index.unique(level=1).tolist()
expanded_indices=list(itertools.product(constituents_list,decades_list))
index=pd.MultiIndex.from_tuples(expanded_indices,names=['constituent','decade'])
```

```python
new_constituents=pd.DataFrame(pd.Series(0,index))
new_constituents.drop(0,axis=1,inplace=True)
constituents=pd.merge(new_constituents.reset_index(),constituents.reset_index(),on=['constituent','decade'],how='left')
constituents['word']=constituents['constituent']+"_"+constituents['decade'].astype(str)
constituents.set_index(['constituent','decade'],inplace=True)
constituents
```
